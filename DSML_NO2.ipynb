```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv("/content/Mall_Customers.csv")
```

```python
dataset.head()
```

```python
dataset.describe()
```

```python
df = pd.DataFrame(dataset)
print(df.var())
```

```python
print(df.dtypes)
```

```python
print(range)
for data in dataset:
  if(data != 'Genre'):
      range= df[data].max()-df[data].min()
      print(data ," : ", range)
```

```python
print(df.shape)
```

```python
for column in df.columns:
    plt.figure(figsize=(4, 3))
    sns.histplot(dataset[column], kde=True,bins=50)
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```
